```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bank = pd.read_pickle("all_banks")
```

```python
bank.head()
```

**What is the max close price for each bank stock throughout the period?**

```python
bank.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
for tick in tickers:
    print(tick, bank[tick]['Close'].max())
```

**Create a new DataFrame called 'returns'. This DataFrame will contain the return for each bank stocks**

```python
returns = pd.DataFrame()
```


**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns [tick + ' Return'] = bank[tick]['Close'].pct_change()
returns.head()         
```

**Create a pairplot for the returns of stock**

```python
sns.pairplot(data=returns)
```

C stock returns stand out to me. 

**Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. You should notice that 4 of the banks share the same day for the worst drop, did anything significant happen that day?**

```python
returns.idxmax()
```

```python
returns.idxmin()
```

Bank of America, Goldman Sachs, JP Morgan Chase, and Wells Fargo had the worst single return on 2009-01-20 as Barack Obama was inaugurated as 44th President of America.

**Create displot for 2015 Morgan Stanley returns**

```python
plt.figure(figsize=(9,6))
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], bins=100)
```

```python
plt.figure(figsize=(9,6))
sns.distplot(returns.loc['2008-01-01':'2008-12-31']['C Return'], color="green", bins=100)
```

## **Moving Averages** ##

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12,9))
bank['BAC']['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30-day moving average')
bank['BAC']['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC close')
plt.legend()
```

**Create a heatmap of the correlation between the stocks Close Price**

```python
sns.heatmap(bank.xs(key='Close', axis =1, level = 'Stock Info').corr(), annot=True)
```

**Visualizing the correlation between bank closing price using cufflinks**

```python
close_corr = bank.xs(key='Close', axis=1, level='Stock Info').corr()
close_corr.iplot(kind='heatmap', colorscale = 'rdylbu')
```
